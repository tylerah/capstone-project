```python
# Creating Race Columns & manual encoding
data['white'] = data['q5'].apply(lambda x: 1 if 'E' in str(x) else 0)
data['AIN'] = data['q5'].apply(lambda x: 1 if 'A' in str(x) else 0)
data['asian'] = data['q5'].apply(lambda x: 1 if 'B' in str(x) else 0)
data['black'] = data['q5'].apply(lambda x: 1 if 'C' in str(x) else 0)
data['PI'] = data['q5'].apply(lambda x: 1 if 'D' in str(x) else 0)

# Dropping q5
data = data.drop(data[["q5"]], axis=1)
data
```

```python
# Changing q28 column 
data.loc[data['q28'] > 0, 'q28'] = 1
```
